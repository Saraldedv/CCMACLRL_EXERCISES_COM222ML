## Building Different Regression Models

### Model 1: Miles Traveled and Travel Time

16. Set independent and dependent variables

```python
# put your answer here
IV = [sm.add_constant(df["Miles Traveled"]), sm.add_constant(df["Number of Deliveries"]), sm.add_constant(df["Gas Price"])]
X = [IV[0], IV[1], IV[2]]
y = df["Travel Time"]
```

17. Build and train the model

```python
# put your answer here
LR = LinearRegression()

res = [LR.fit(i, y) for i in X]
```

18. Display the statistical summary

```python
# put your answer here
olsmod = [sm.OLS(y, i).fit() for i in X]
olsmod[0].summary()
```

19. Display the r-squared

```python
# put your answer here
olsmod_rsquare = [i.rsquared for i in olsmod]

x_axis = ["Miles Traveled", "Number of Deliveries", "Gas Price"]
y_axis = olsmod_rsquare

plt.bar(x_axis, y_axis)
plt.title("R Square Per Independent Variable")
plt.xlabel("Iddependent variables")
plt.ylabel("r_squared")
plt.show()
```
